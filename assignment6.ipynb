## 6.1 Overview of Data Engineering in Applied Data Science

### Q1. What is data engineering? Explain the role of a data engineer in a data science project. How does it differ from data science? [10 marks]

**Data Engineering** is the discipline focused on designing, building, and maintaining the infrastructure, pipelines, and systems that collect, store, process, and deliver data reliably and at scale — so that data analysts and data scientists can work with clean, timely, and accessible data.

Data engineering is the *plumbing* of the data ecosystem: without it, no amount of sophisticated machine learning models can function reliably in production.

---

#### Role of a Data Engineer

| Responsibility | Description |
|---|---|
| **Pipeline development** | Build ETL/ELT pipelines to move data from source systems to storage/analytics layers |
| **Data infrastructure** | Set up and manage databases, data warehouses, data lakes, and streaming systems |
| **Data quality** | Implement validation, deduplication, and cleansing logic to ensure data reliability |
| **Scalability & performance** | Optimize queries, partitioning, and distributed processing for large datasets |
| **Orchestration** | Schedule and monitor pipeline jobs using tools like Airflow, Prefect, or CRON |
| **Security & governance** | Enforce access controls, encryption, and data cataloguing |
| **Collaboration** | Work with data scientists to understand data requirements; expose clean datasets via APIs or SQL |

---

#### Data Engineering vs Data Science

| Dimension | Data Engineer | Data Scientist |
|---|---|---|
| **Primary goal** | Build and maintain data infrastructure | Extract insights and build predictive models |
| **Focus** | Data availability, reliability, scale | Data analysis, statistics, machine learning |
| **Output** | Pipelines, databases, APIs, data products | Models, dashboards, reports, recommendations |
| **Languages/Tools** | Python, SQL, Spark, Kafka, Airflow, Docker | Python, R, TensorFlow, Jupyter, Tableau |
| **Skills** | Software engineering, distributed systems, SQL | Statistics, ML, communication, domain knowledge |
| **Works with** | Raw, messy, high-volume data | Clean, processed, analysis-ready data |
| **End user** | Data scientists, analysts, applications | Business stakeholders, product teams |

**Analogy:** The data engineer builds and maintains the highway; the data scientist drives on it to reach insights.

In production systems, both roles are tightly coupled: data scientists specify what data they need; data engineers build the pipelines to deliver it reliably. Modern "full-stack" data teams often expect overlap in these roles.

### Q2. Describe the overall data engineering lifecycle. Key stages from ingestion to serving. [10 marks]

The **Data Engineering Lifecycle** is the end-to-end process through which raw data from source systems is transformed into analytics-ready information.

```
  ┌──────────┐    ┌──────────┐    ┌──────────────┐    ┌──────────┐    ┌──────────┐
  │ SOURCE   │───▶│ INGEST   │───▶│   STORE      │───▶│ PROCESS  │───▶│  SERVE   │
  │ SYSTEMS  │    │          │    │              │    │          │    │          │
  └──────────┘    └──────────┘    └──────────────┘    └──────────┘    └──────────┘
  DBs, APIs,      Batch/Stream    Data Lake /          Transform,      BI tools,
  Files, IoT      pipelines       Data Warehouse        Clean, Join    ML models
```

---

#### Stage 1: Data Generation & Sources
- Transactional databases (MySQL, PostgreSQL)
- REST APIs, web scraping
- IoT sensors, log files, clickstream data
- SaaS applications (Salesforce, Google Analytics)

#### Stage 2: Ingestion
Two patterns:
- **Batch ingestion:** Periodic bulk transfer (e.g., nightly CSV dump from ERP) — lower latency requirements, simpler
- **Stream ingestion:** Continuous real-time data flow (Apache Kafka, AWS Kinesis) — for use cases requiring near-instant data (fraud detection, real-time dashboards)

#### Stage 3: Storage
| Storage Type | Characteristics | Use Case |
|---|---|---|
| **Data Lake** | Raw, schema-on-read, cheap (S3/HDFS), any format | Store everything; ML feature stores |
| **Data Warehouse** | Structured, schema-on-write, optimized for SQL analytics (Snowflake, BigQuery) | BI reporting, aggregations |
| **Data Lakehouse** | Combines lake flexibility + warehouse ACID (Delta Lake, Apache Iceberg) | Modern unified analytics |
| **Operational DB** | OLTP, transactional, row-oriented | Real-time application storage |

#### Stage 4: Processing / Transformation
- **Batch processing:** Apache Spark, dbt, Pandas — process large datasets in scheduled windows
- **Stream processing:** Apache Flink, Spark Streaming — process events as they arrive
- Operations: cleaning, deduplication, joins, aggregations, feature engineering

#### Stage 5: Serving
- **BI/Analytics:** Tableau, Power BI, Metabase connect to data warehouse
- **ML serving:** Feature store exposes features to model serving APIs
- **Data APIs:** REST/GraphQL endpoints serve processed data to applications
- **Reports:** Scheduled automated report generation (Excel, PDF, HTML)

#### Stage 6: Orchestration & Monitoring (cross-cutting)
- Apache Airflow manages task dependencies, scheduling, retries
- Monitoring: data quality checks, pipeline SLA tracking, alerting on failures
- Lineage: track data provenance from source to destination

### Q3. Differentiate between structured, semi-structured, and unstructured data. How does each type affect data engineering design decisions? [5 marks]

| Feature | Structured | Semi-Structured | Unstructured |
|---|---|---|---|
| **Definition** | Highly organized, fits in a predefined relational schema | Has some organizational structure but no rigid schema | No predefined format or organization |
| **Examples** | SQL tables, CSV, Excel | JSON, XML, YAML, Avro, Parquet | Images, videos, audio, PDFs, log files, social media text |
| **Schema** | Fixed (schema-on-write) | Flexible (self-describing) | None |
| **Query language** | SQL | JSONPath, XPath, or SQL with JSON functions | Full-text search, ML-based |
| **Storage** | RDBMS (MySQL, PostgreSQL) | Document stores (MongoDB), object storage | Object storage (S3), blob storage |
| **% of world's data** | ~20% | ~30% | ~50% |

**Impact on Data Engineering Design Decisions:**

1. **Structured:** Use RDBMS or columnar data warehouses. Schema migrations must be versioned. ETL is straightforward — SQL transformations, JOIN operations. Best fit for reporting and BI.

2. **Semi-structured:** Requires schema evolution handling (new JSON fields can appear). Use document databases (MongoDB) or schema-on-read data lakes. Parser logic must handle missing keys gracefully. Common in REST API responses and event streams.

3. **Unstructured:** Cannot be queried with SQL directly. Requires preprocessing pipelines (OCR for PDFs, feature extraction for images). Storage in blob/object stores. Processing requires ML models or specialized tools (OpenCV, NLP frameworks). Most expensive to process and store per insight.

### Q4. Difference between data warehouse, data lake, and data lakehouse. When to use each? [5 marks]

| Feature | Data Warehouse | Data Lake | Data Lakehouse |
|---|---|---|---|
| **Schema** | Schema-on-write (defined before loading) | Schema-on-read (defined at query time) | Both — schema-on-write for curated zones |
| **Data type** | Structured only | All types (structured, semi, unstructured) | All types |
| **Storage cost** | High (optimized proprietary storage) | Low (commodity object storage: S3, ADLS) | Low-medium |
| **Query performance** | Very fast (pre-optimized) | Slower (no indexing by default) | Fast (with table formats) |
| **ACID transactions** | ✅ Full ACID | ❌ No | ✅ Yes (Delta Lake, Iceberg) |
| **Data quality** | High — validated on ingestion | Raw — may have quality issues | High with governance layer |
| **Best for** | BI reports, aggregations, SQL analytics | ML training data, data exploration, archiving | Modern unified analytics platforms |
| **Tools** | Snowflake, BigQuery, Redshift, SQL Server | AWS S3 + Glue, Azure Data Lake, HDFS | Databricks (Delta Lake), Apache Iceberg |

**When to use each:**
- **Data Warehouse:** Business intelligence, standardized reporting, financial analytics, where data consistency and query speed are paramount. Example: Monthly revenue reporting for the CFO.
- **Data Lake:** When you need to store everything (raw logs, clickstreams, images) for future use cases not yet defined, or for machine learning experiments. Example: Storing all web logs for training a recommendation engine.
- **Data Lakehouse:** Modern choice for organizations wanting the flexibility of a lake with the reliability of a warehouse. Example: Netflix, Airbnb — need both real-time ML and BI from the same platform.

### Q5. Explain data pipeline orchestration. What problems does it solve in production data systems? [5 marks]

**Pipeline Orchestration** is the automated coordination, scheduling, execution monitoring, and error handling of multi-step data workflows — ensuring that tasks run in the correct order, at the right time, with proper handling of failures.

**Problems Solved by Orchestration:**

| Problem Without Orchestration | Orchestration Solution |
|---|---|
| Tasks run in wrong order (Step B starts before Step A finishes) | **Dependency management** — DAG (Directed Acyclic Graph) defines task order |
| Silent failures — pipeline crashes at 3 AM with no notification | **Alerting & monitoring** — sends email/Slack on failure; dashboard shows status |
| No retry logic — transient errors cause permanent data loss | **Automatic retries** — configurable retry count and backoff delay |
| Manual triggering — engineer must SSH to server to run job | **Scheduling** — CRON-like scheduling or event-based triggers |
| Hard to debug failures in 20-step pipeline | **Logging & lineage** — per-task logs, execution history, data lineage tracking |
| Re-running failed jobs processes already-complete steps | **Idempotency + partial re-run** — only re-run failed tasks |
| No visibility into what's running / SLA breaches | **SLA monitoring** — alerts when task exceeds expected duration |

**Key Orchestration Tools:**
- **Apache Airflow:** Most widely used; Python DAGs; web UI; large ecosystem
- **Prefect:** Modern, Pythonic; cloud-native; better local testing
- **Dagster:** Asset-based orchestration; strong typing; great for data engineering
- **CRON + shell scripts:** Simple, OS-level; adequate for single-task jobs

## 6.2 Designing and Implementing ETL Pipelines

### Q1. What is an ETL pipeline? Explain each stage with a real-world example. Implement a simple ETL pipeline using pandas. [10 marks]

**ETL (Extract, Transform, Load)** is a data integration pattern consisting of three sequential stages:

| Stage | Description | Real-World Example |
|---|---|---|
| **Extract** | Read raw data from one or more source systems | Pull daily sales records from a MySQL database and customer data from a REST API |
| **Transform** | Clean, validate, enrich, and reshape the data | Remove duplicate orders, standardize date formats, compute total_revenue = quantity × price, filter cancelled orders |
| **Load** | Write the transformed data to the target system | Write aggregated daily sales summary into a PostgreSQL data warehouse for BI reporting |

````python
import pandas as pd
import sqlite3
import numpy as np
import logging
import os
from datetime import datetime

logging.basicConfig(level=logging.INFO,
                    format='%(asctime)s [%(levelname)s] %(message)s')
logger = logging.getLogger(__name__)

# ── Generate sample source CSV ─────────────────────────────────
np.random.seed(42)
n = 200
raw_df = pd.DataFrame({
    'order_id'   : range(1001, 1001 + n),
    'customer'   : np.random.choice(['Alice','Bob','Carol','Dave','Eve'], n),
    'product'    : np.random.choice(['Laptop','Phone','Tablet','Watch'], n),
    'quantity'   : np.random.randint(1, 10, n),
    'unit_price' : np.random.uniform(500, 5000, n).round(2),
    'order_date' : pd.date_range('2024-01-01', periods=n, freq='D').strftime('%d/%m/%Y'),
    'status'     : np.random.choice(['completed','cancelled','pending'], n, p=[0.7,0.2,0.1])
})
# Inject data quality issues
raw_df.loc[np.random.choice(n, 15, replace=False), 'unit_price'] = np.nan
raw_df.loc[np.random.choice(n, 8, replace=False), 'quantity']   = np.nan
raw_df = pd.concat([raw_df, raw_df.sample(10)], ignore_index=True)  # duplicates
raw_df.to_csv('raw_sales.csv', index=False)
logger.info("Sample CSV created: raw_sales.csv")


# ═══════════════════════════════════════════════════════════════
# EXTRACT
# ═══════════════════════════════════════════════════════════════
def extract(filepath):
    """Read raw data from CSV source."""
    logger.info(f"[EXTRACT] Reading from: {filepath}")
    df = pd.read_csv(filepath)
    logger.info(f"[EXTRACT] Loaded {len(df)} rows, {df.shape[1]} columns")
    return df


# ═══════════════════════════════════════════════════════════════
# TRANSFORM
# ═══════════════════════════════════════════════════════════════
def transform(df):
    """Clean, validate, and enrich the raw data."""
    logger.info("[TRANSFORM] Starting transformation...")
    original_rows = len(df)

    # Step 1: Remove duplicates
    df = df.drop_duplicates(subset=['order_id'])
    logger.info(f"[TRANSFORM] Removed {original_rows - len(df)} duplicate rows")

    # Step 2: Handle missing values
    null_before = df.isnull().sum().sum()
    df['unit_price'].fillna(df['unit_price'].median(), inplace=True)
    df['quantity'].fillna(1, inplace=True)
    logger.info(f"[TRANSFORM] Imputed {null_before} missing values")

    # Step 3: Standardize date format
    df['order_date'] = pd.to_datetime(df['order_date'], format='%d/%m/%Y')

    # Step 4: Type casting
    df['quantity']   = df['quantity'].astype(int)
    df['unit_price'] = df['unit_price'].round(2)

    # Step 5: Filter — keep only completed orders
    before_filter = len(df)
    df = df[df['status'] == 'completed'].copy()
    logger.info(f"[TRANSFORM] Filtered to completed orders: {before_filter} → {len(df)} rows")

    # Step 6: Feature engineering
    df['total_revenue'] = (df['quantity'] * df['unit_price']).round(2)
    df['order_month']   = df['order_date'].dt.to_period('M').astype(str)

    logger.info(f"[TRANSFORM] Complete. Output: {len(df)} rows, {df.shape[1]} columns")
    return df


# ═══════════════════════════════════════════════════════════════
# LOAD
# ═══════════════════════════════════════════════════════════════
def load(df, db_path='pipeline.db', table='sales_clean'):
    """Load transformed data into SQLite database."""
    logger.info(f"[LOAD] Writing {len(df)} rows to '{table}' in {db_path}")
    conn = sqlite3.connect(db_path)
    try:
        df.to_sql(table, conn, if_exists='replace', index=False)
        logger.info(f"[LOAD] Successfully loaded {len(df)} rows")
    finally:
        conn.close()


# ═══════════════════════════════════════════════════════════════
# RUN ETL
# ═══════════════════════════════════════════════════════════════
raw      = extract('raw_sales.csv')
clean    = transform(raw)
load(clean)

# Verify
conn = sqlite3.connect('pipeline.db')
result = pd.read_sql('SELECT product, SUM(total_revenue) as revenue FROM sales_clean GROUP BY product ORDER BY revenue DESC', conn)
conn.close()
print("\nRevenue by Product:")
print(result.round(2).to_string(index=False))
````

### Q2. Differentiate between ETL and ELT. When is ELT preferred? Examples with modern cloud tools. [10 marks]

| Dimension | ETL (Extract → Transform → Load) | ELT (Extract → Load → Transform) |
|---|---|---|
| **Transform location** | On a separate ETL server/tool BEFORE loading | Inside the target database/warehouse AFTER loading |
| **Data loaded** | Only clean, transformed data | Raw data is loaded first |
| **Processing power** | ETL server must be powerful | Uses the massive compute of cloud warehouses |
| **Storage cost** | Lower (only clean data stored) | Higher (raw + processed both stored) |
| **Flexibility** | Lower — schema must be defined upfront | Higher — raw data available for re-transformation |
| **Latency** | Higher — transformation bottleneck | Lower — load fast, transform in parallel |
| **Best for** | Traditional on-premise DWH, sensitive data that shouldn't be stored raw | Cloud-scale data warehouses, big data, ML pipelines |
| **Tools** | Informatica, Talend, SSIS, Pandas pipelines | Snowflake + dbt, BigQuery + dbt, Databricks |

**When is ELT Preferred?**

1. **Cloud-scale data volumes:** Cloud warehouses (Snowflake, BigQuery, Redshift) have massive parallel processing — transforming 1TB inside BigQuery is faster than on a dedicated ETL server.

2. **Schema flexibility needed:** When source data schema changes frequently, storing raw JSON in the lake and transforming on-demand is more resilient than re-engineering an ETL schema.

3. **Multiple downstream consumers:** Raw data stored in the lake can be transformed differently by different teams (BI team wants aggregates; ML team wants features) without moving data twice.

4. **Audit trail requirements:** Storing raw data enables re-processing with corrected logic if a transformation bug is discovered.

5. **Modern dbt workflows:** dbt (data build tool) enables ELT by running SQL transformations directly inside the warehouse with version control, testing, and documentation.

**Modern Cloud ELT Example:**
```
Source DB → Fivetran (Extract + Load) → Snowflake Raw Layer → dbt (Transform) → Analytics Layer → Tableau
```

**Traditional On-Premise ETL Example:**
```
Source DB → Informatica ETL Server (Extract + Transform) → SQL Server DWH (Load) → SSRS Reports
```

### Q3. Python ETL pipeline: extract CSV → clean/rename → filter → load to SQLite. [10 marks]

````python
import pandas as pd
import sqlite3
import numpy as np

# ── Generate a sample CSV (employee dataset) ──────────────────
np.random.seed(10)
n = 150
emp_df = pd.DataFrame({
    'EMP_ID'    : range(101, 101 + n),
    'FULL_NAME' : [f'Employee_{i}' for i in range(n)],
    'DEPT'      : np.random.choice(['  IT', 'HR ', ' Finance', 'Marketing '], n),
    'SALARY'    : np.random.randint(30000, 120000, n).astype(float),
    'AGE'       : np.random.randint(22, 60, n),
    'JOIN_DATE' : pd.date_range('2015-01-01', periods=n, freq='W').strftime('%Y-%m-%d')
})
# Inject issues
emp_df.loc[np.random.choice(n, 12, replace=False), 'SALARY'] = np.nan
emp_df.loc[np.random.choice(n, 5, replace=False), 'AGE']    = np.nan
emp_df = pd.concat([emp_df, emp_df.sample(8)], ignore_index=True)
emp_df.to_csv('employees.csv', index=False)
print(f"Source CSV created: employees.csv ({len(emp_df)} rows)")


# ═══════════════════════════════════════════════════════════════
# (i) EXTRACT — Read from CSV
# ═══════════════════════════════════════════════════════════════
def extract(path):
    print(f"\n{'─'*50}")
    print("[EXTRACT] Reading CSV...")
    df = pd.read_csv(path)
    print(f"  Loaded: {df.shape[0]} rows × {df.shape[1]} columns")
    print(f"  Columns: {df.columns.tolist()}")
    print(f"  Missing values: {df.isnull().sum().to_dict()}")
    return df


# ═══════════════════════════════════════════════════════════════
# (ii) TRANSFORM — Clean, rename, fix formats
# ═══════════════════════════════════════════════════════════════
def transform(df):
    print(f"\n{'─'*50}")
    print("[TRANSFORM] Cleaning and transforming...")

    # Rename columns — lowercase, underscores
    df = df.rename(columns={
        'EMP_ID'   : 'employee_id',
        'FULL_NAME': 'employee_name',
        'DEPT'     : 'department',
        'SALARY'   : 'salary',
        'AGE'      : 'age',
        'JOIN_DATE': 'join_date'
    })
    print("  ✅ Columns renamed")

    # Remove duplicates
    before = len(df)
    df = df.drop_duplicates(subset=['employee_id'])
    print(f"  ✅ Removed {before - len(df)} duplicate rows")

    # Strip whitespace from string columns
    df['department'] = df['department'].str.strip().str.upper()
    df['employee_name'] = df['employee_name'].str.strip()
    print("  ✅ Stripped whitespace from string columns")

    # Handle missing values
    df['salary'].fillna(df['salary'].median(), inplace=True)
    df['age'].fillna(df['age'].median().astype(int), inplace=True)
    df['age'] = df['age'].astype(int)
    print(f"  ✅ Imputed missing salary and age values")

    # Parse dates
    df['join_date'] = pd.to_datetime(df['join_date'])
    df['years_at_company'] = (pd.Timestamp.today() - df['join_date']).dt.days // 365
    print("  ✅ Parsed join_date and computed years_at_company")

    # Feature: salary band
    df['salary_band'] = pd.cut(df['salary'],
                               bins=[0, 50000, 80000, 120000],
                               labels=['Junior', 'Mid', 'Senior'])
    print(f"  ✅ Created salary_band feature")
    print(f"  Output: {df.shape[0]} rows × {df.shape[1]} columns")
    return df


# ═══════════════════════════════════════════════════════════════
# (iii) FILTER — Only employees with salary > 50,000
# ═══════════════════════════════════════════════════════════════
def filter_data(df, min_salary=50000):
    before = len(df)
    df = df[df['salary'] > min_salary].copy()
    print(f"\n[FILTER] salary > {min_salary:,}: {before} → {len(df)} rows")
    return df


# ═══════════════════════════════════════════════════════════════
# (iv) LOAD — Write to SQLite
# ═══════════════════════════════════════════════════════════════
def load(df, db='hr_pipeline.db', table='employees_clean'):
    print(f"\n{'─'*50}")
    print(f"[LOAD] Writing {len(df)} rows to SQLite → table: {table}")

    df_save = df.copy()
    df_save['join_date']   = df_save['join_date'].astype(str)
    df_save['salary_band'] = df_save['salary_band'].astype(str)

    conn = sqlite3.connect(db)
    try:
        df_save.to_sql(table, conn, if_exists='replace', index=False)
        print(f"  ✅ Loaded successfully into '{db}'")
    except Exception as e:
        print(f"  ❌ Load failed: {e}")
        raise
    finally:
        conn.close()


# ═══════════════════════════════════════════════════════════════
# RUN PIPELINE
# ═══════════════════════════════════════════════════════════════
raw_data       = extract('employees.csv')
clean_data     = transform(raw_data)
filtered_data  = filter_data(clean_data, min_salary=50000)
load(filtered_data)

# Verify results
print(f"\n{'─'*50}")
print("[VERIFY] Sample query from SQLite:")
conn = sqlite3.connect('hr_pipeline.db')
summary = pd.read_sql(
    "SELECT department, COUNT(*) as headcount, ROUND(AVG(salary),0) as avg_salary "
    "FROM employees_clean GROUP BY department ORDER BY avg_salary DESC", conn
)
conn.close()
print(summary.to_string(index=False))
````

### Q4. Common data quality issues in ETL pipelines. How to handle missing values, duplicates, and inconsistent formats? [5 marks]

**Common Data Quality Issues and Remediation:**

| Issue | Description | Pandas Remediation |
|---|---|---|
| **Missing values** | NaN, None, empty strings in numerical/categorical fields | `df.fillna(df.median())` for numerical; `df.fillna(df.mode()[0])` for categorical; `df.dropna(subset=['key_col'])` for critical fields |
| **Duplicates** | Same record appearing multiple times due to repeated API calls or join errors | `df.drop_duplicates()` for full row; `df.drop_duplicates(subset=['id'])` for key-based |
| **Inconsistent formats** | Dates as `'2024/01/15'` vs `'15-01-2024'`; strings with extra spaces | `pd.to_datetime(df['date'], infer_datetime_format=True)` |
| **Wrong data types** | Age stored as string `'25'`, price stored as object `'$500'` | `df['age'].astype(int)`; `df['price'].str.replace('$','').astype(float)` |
| **Outliers / invalid values** | Age = 999, negative prices, future dates in past events | `df = df[df['age'].between(0, 120)]`; clip: `df['price'].clip(lower=0)` |
| **Inconsistent casing** | `'new york'`, `'New York'`, `'NEW YORK'` in same column | `df['city'].str.strip().str.title()` |
| **Referential integrity** | Foreign key values that don't exist in parent table | Merge with parent, identify unmatched rows with `indicator=True` |
| **Schema drift** | New columns appear in source data unexpectedly | Validate against expected schema; log warnings for new columns |

**Best Practice — Data Quality Check Function:**

### Q5. Explain incremental loading in ETL. How does it differ from full load? Why preferred for large datasets? [5 marks]

| Feature | Full Load | Incremental Load |
|---|---|---|
| **Definition** | Re-processes and re-loads the **entire dataset** every run | Processes and loads **only new or changed records** since the last successful run |
| **Data processed per run** | 100% of all historical data | Only delta (new/updated/deleted rows) |
| **Processing time** | Grows linearly with total data size | Approximately constant per run (proportional to change volume) |
| **Resource usage** | High CPU, memory, I/O, network | Low — proportional to change rate |
| **Complexity** | Simple — truncate and reload | Higher — needs watermarks, change detection |
| **Risk** | Safe — source of truth always applied | Risk of missing changes if watermark logic has bugs |
| **Use case** | Small tables, reference/lookup data, initial load | Large transactional tables, event logs, time-series data |

**Incremental Loading Mechanisms:**

1. **Timestamp-based watermark:** Query `WHERE updated_at > last_run_timestamp`. Simple and common.
   ```python
   last_run = '2024-06-01'
   new_records = pd.read_sql(
       f"SELECT * FROM orders WHERE updated_at > '{last_run}'", conn
   )
   ```

2. **Auto-increment ID watermark:** `WHERE order_id > last_max_id`. Works when IDs are monotonically increasing.

3. **CDC (Change Data Capture):** Reads database transaction log (binlog in MySQL, WAL in PostgreSQL) to capture all inserts/updates/deletes. Most accurate but complex. Tools: Debezium, AWS DMS.

**Why preferred for large datasets:**
- A full load of 500GB table nightly → costs hours of processing and high cloud compute bills
- Incremental load of 500MB daily change → completes in minutes at a fraction of the cost
- Reduces load on source systems (fewer queries hitting production DB)
- Enables near-real-time data freshness when combined with streaming

### Q6. Best practices for designing a robust and scalable ETL pipeline. [5 marks]

**1. Idempotency**
Re-running the pipeline should produce the same result as running it once. Achieved by using `if_exists='replace'` in SQL loads, or upsert (INSERT OR REPLACE) logic instead of plain INSERT.
- *Without idempotency:* Re-running a failed pipeline doubles records or corrupts aggregations

**2. Modularity**
Separate extract, transform, and load into independent functions/classes. Each function has a single responsibility, is independently testable, and can be replaced without touching others.
```python
def extract(): ...
def transform(df): ...
def load(df): ...
def run_pipeline(): load(transform(extract()))
```

**3. Comprehensive Logging**
Log at every stage: rows extracted, rows after cleaning, rows loaded, execution time. Log warnings for data quality issues. Write logs to both console and rotating file. Use structured (JSON) logging in production.

**4. Error Handling**
Wrap every stage in try-except. Distinguish between retriable errors (network timeout → retry) and fatal errors (schema mismatch → alert and halt). Use a dead-letter mechanism to capture and store failed records.

**5. Data Validation**
After each stage, assert invariants: row count > 0, no nulls in primary key column, value ranges within expected bounds. Tools: Great Expectations, Pandera.

**6. Incremental + Watermark Design**
Avoid full-table scans where possible. Persist the last successful watermark to a state store.

**7. Configuration-Driven Design**
Externalize file paths, DB credentials, thresholds to config files or environment variables — never hardcode.

### Q1. What is CRON? Explain expression syntax with examples. Write 4 CRON expressions. [10 marks]

**CRON** is a Unix/Linux time-based job scheduler daemon that executes commands or scripts at specified times. It reads a configuration file called `crontab` (CRON table) containing scheduled tasks.

**CRON Expression Syntax:**

```
  ┌─────────────── minute       (0–59)
  │  ┌──────────── hour         (0–23)
  │  │  ┌───────── day of month (1–31)
  │  │  │  ┌────── month        (1–12)
  │  │  │  │  ┌─── day of week  (0–7, 0 and 7 = Sunday)
  │  │  │  │  │
  *  *  *  *  *   command_to_execute
```

**Special Characters:**
| Symbol | Meaning | Example |
|---|---|---|
| `*` | Every possible value | `* * * * *` → every minute |
| `,` | List of values | `0 9,17 * * *` → 9 AM and 5 PM daily |
| `-` | Range | `0 9-17 * * 1-5` → hourly 9 AM–5 PM weekdays |
| `*/n` | Every n-th value | `*/15 * * * *` → every 15 minutes |
| `@` | Special macros | `@daily`, `@weekly`, `@monthly`, `@reboot` |

---

**Required CRON Expressions:**

| Task | CRON Expression | Explanation |
|---|---|---|
| **(i) Every day at midnight** | `0 0 * * *` | minute=0, hour=0, every day |
| **(ii) Every Monday at 9 AM** | `0 9 * * 1` | minute=0, hour=9, weekday=1 (Monday) |
| **(iii) Every 15 minutes** | `*/15 * * * *` | every 15 minutes, all hours, all days |
| **(iv) First day of every month** | `0 0 1 * *` | minute=0, hour=0, day=1, any month |

**Additional examples:**
```bash
# Every weekday at 8 AM
0 8 * * 1-5  /path/to/etl_pipeline.py

# Every 30 minutes between 9 AM and 6 PM
*/30 9-18 * * *  /path/to/script.py

# Last day of month (workaround)
0 23 28-31 * *  [ "$(date +\%d -d tomorrow)" = "01" ] && /path/to/script.py
```

**How to set up a CRON job:**
```bash
crontab -e          # Edit crontab for current user
crontab -l          # List all CRON jobs
crontab -r          # Remove all CRON jobs

# Example entry:
0 8 * * 1-5 /usr/bin/python3 /home/user/etl/pipeline.py >> /var/log/etl.log 2>&1
```

````python
import schedule
import time
import logging
import pandas as pd
import numpy as np
import os
from datetime import datetime

# ── Logging Setup ─────────────────────────────────────────────
os.makedirs('logs', exist_ok=True)

logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s [%(levelname)s] %(name)s — %(message)s',
    handlers=[
        logging.FileHandler('logs/pipeline.log'),       # Write to file
        logging.StreamHandler()                         # Print to console
    ]
)
logger = logging.getLogger('DataPipeline')


# ── Pipeline Functions ─────────────────────────────────────────
def generate_hourly_data():
    """Simulate hourly data arriving in a CSV."""
    n = np.random.randint(20, 80)
    df = pd.DataFrame({
        'timestamp': pd.date_range(datetime.now(), periods=n, freq='T').astype(str),
        'sensor_id': np.random.choice(['S1','S2','S3'], n),
        'value'    : np.random.normal(50, 10, n).round(2),
        'status'   : np.random.choice(['ok','warn','error'], n, p=[0.8,0.15,0.05])
    })
    # Inject some missing values
    df.loc[np.random.choice(n, 3, replace=False), 'value'] = np.nan
    df.to_csv('source_data.csv', index=False)
    return n


def extract_step():
    logger.info("[EXTRACT] Reading source_data.csv")
    try:
        df = pd.read_csv('source_data.csv')
        logger.info(f"[EXTRACT] Loaded {len(df)} records")
        return df
    except FileNotFoundError:
        logger.error("[EXTRACT] source_data.csv not found!")
        return None


def process_step(df):
    if df is None:
        return None
    logger.info("[PROCESS] Cleaning and aggregating...")

    null_count = df['value'].isnull().sum()
    if null_count > 0:
        logger.warning(f"[PROCESS] {null_count} null values in 'value' column — imputing with median")
        df['value'].fillna(df['value'].median(), inplace=True)

    df_ok = df[df['status'] == 'ok']
    summary = df_ok.groupby('sensor_id').agg(
        count=('value', 'count'),
        mean_value=('value', 'mean'),
        max_value=('value', 'max'),
        min_value=('value', 'min')
    ).round(2).reset_index()
    summary['run_time'] = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
    logger.info(f"[PROCESS] Aggregated {len(df_ok)} ok records into {len(summary)} sensor summaries")
    return summary


def save_step(df):
    if df is None:
        return
    out_path = 'processed_output.csv'
    # Append to existing output file
    if os.path.exists(out_path):
        existing = pd.read_csv(out_path)
        df = pd.concat([existing, df], ignore_index=True)
    df.to_csv(out_path, index=False)
    logger.info(f"[SAVE] Output written to {out_path} ({len(df)} total rows)")


# ── Main Pipeline ──────────────────────────────────────────────
def run_pipeline():
    start = datetime.now()
    logger.info(f"{'='*50}")
    logger.info(f"Pipeline run started at {start.strftime('%Y-%m-%d %H:%M:%S')}")

    try:
        n = generate_hourly_data()
        raw  = extract_step()
        proc = process_step(raw)
        save_step(proc)

        elapsed = (datetime.now() - start).total_seconds()
        logger.info(f"Pipeline completed in {elapsed:.2f}s")

    except Exception as e:
        logger.error(f"Pipeline FAILED: {e}", exc_info=True)


# ── Scheduling ─────────────────────────────────────────────────
# In production: schedule.every().hour.do(run_pipeline)
# For demonstration, run every 10 seconds and show 3 iterations

schedule.every(10).seconds.do(run_pipeline)   # Change to .hour for production
# schedule.every().hour.do(run_pipeline)       # Every hour
# schedule.every().day.at('08:00').do(run_pipeline)  # Daily at 8 AM
# schedule.every().monday.at('09:00').do(run_pipeline)  # Every Monday 9 AM

logger.info("Scheduler started. Running pipeline every 10 seconds (demo mode).")
logger.info("In production, change to: schedule.every().hour.do(run_pipeline)")

MAX_RUNS = 3
run_count = 0

while run_count < MAX_RUNS:
    schedule.run_pending()
    time.sleep(1)
    # Count completed jobs
    if not schedule.jobs[0].should_run:
        pass
    # Simple counter approach
    if schedule.idle_seconds() <= 0:
        pass

# Alternative: manual demo run for notebook
schedule.clear()
logger.info("\nDirect demo run:")
for i in range(2):
    logger.info(f"--- Iteration {i+1} ---")
    run_pipeline()
    time.sleep(1)

print("\nFinal processed output:")
print(pd.read_csv('processed_output.csv'))
````

### Q4. What is Apache Airflow? How does it improve upon CRON? Explain DAGs. [5 marks]

**Apache Airflow** is an open-source workflow orchestration platform that allows you to programmatically author, schedule, and monitor data pipelines as code (Python). Created by Airbnb (2014), now an Apache top-level project.

**DAG (Directed Acyclic Graph):**
- **Directed:** Edges have a direction (task A → task B means B depends on A)
- **Acyclic:** No circular dependencies (no loops allowed)
- **Graph:** A collection of tasks (nodes) and their dependencies (edges)

                                   [send_alert_email]

### Q5. Explain workflow orchestration. Problems without it in multi-step data pipelines. [5 marks]

**Workflow Orchestration** is the automated management of multi-step computational workflows — ensuring tasks execute in the correct order, handling failures gracefully, and providing visibility into the execution state of each step.

**Problems Without Proper Orchestration:**

1. **No dependency management:** Step 3 might start before Step 2 has written its output file, causing cryptic "file not found" errors. Developer must manually sequence scripts or add `sleep` delays.

2. **Silent failures:** A pipeline step fails at 3 AM; the downstream steps either crash or produce empty outputs. Nobody notices until business users complain about stale dashboards the next morning.

3. **No retry logic:** A transient network error causes a database connection to fail. Without retries, the entire pipeline halts and must be manually restarted.

4. **Hard to debug:** In a 15-step pipeline using raw CRON jobs, when something fails you must SSH into the server, search through scattered log files, and reconstruct which step failed and why.

5. **No backfill capability:** Server was down for 3 days. Without orchestration, there is no systematic way to re-run only the missed pipeline runs for those days.

6. **No SLA enforcement:** A critical report pipeline that usually finishes in 10 minutes starts taking 3 hours due to a data volume spike — without orchestration, nobody is alerted.

7. **Tight coupling:** Scripts call each other with hardcoded paths and assumptions. Changing one script requires understanding and modifying all its callers.

**Orchestration solves all of these** by providing: DAG-based ordering, per-task logging, automatic retries, email/Slack alerting, backfill, and a unified monitoring dashboard.

### Q2. Error handling in data pipelines: try-except-finally, custom exceptions, retry logic. [10 marks]

````python
import pandas as pd
import sqlite3
import time
import logging
import functools

logger = logging.getLogger('ErrorHandlingDemo')
if not logger.handlers:
    logging.basicConfig(level=logging.INFO,
                        format='%(asctime)s [%(levelname)s] %(message)s')


# ── Custom Exception Hierarchy ─────────────────────────────────
class PipelineError(Exception):
    """Base exception for all pipeline errors."""
    pass

class ExtractionError(PipelineError):
    """Raised when data extraction fails."""
    pass

class TransformationError(PipelineError):
    """Raised when data transformation fails validation."""
    pass

class LoadError(PipelineError):
    """Raised when data loading fails."""
    pass

class DataQualityError(PipelineError):
    """Raised when output data fails quality checks."""
    pass


# ── Retry Decorator ────────────────────────────────────────────
def retry(max_attempts=3, delay_seconds=2, exceptions=(Exception,)):
    """
    Decorator for automatic retry with exponential backoff.
    
    Parameters:
        max_attempts : maximum number of attempts
        delay_seconds: initial wait between retries (doubled each attempt)
        exceptions   : tuple of exception types to retry on
    """
    def decorator(func):
        @functools.wraps(func)
        def wrapper(*args, **kwargs):
            attempt = 0
            last_exception = None
            while attempt < max_attempts:
                try:
                    return func(*args, **kwargs)
                except exceptions as e:
                    attempt += 1
                    last_exception = e
                    wait = delay_seconds * (2 ** (attempt - 1))  # Exponential backoff
                    logger.warning(
                        f"[RETRY] {func.__name__} failed (attempt {attempt}/{max_attempts}): {e}"
                        f" — retrying in {wait}s"
                    )
                    if attempt < max_attempts:
                        time.sleep(wait)
            logger.error(f"[RETRY] {func.__name__} failed after {max_attempts} attempts")
            raise last_exception
        return wrapper
    return decorator


# ── Pipeline Functions with Full Error Handling ────────────────
@retry(max_attempts=3, delay_seconds=1, exceptions=(FileNotFoundError, IOError))
def safe_extract(filepath):
    logger.info(f"[EXTRACT] Attempting to read: {filepath}")
    try:
        df = pd.read_csv(filepath)
        if df.empty:
            raise ExtractionError(f"File {filepath} is empty")
        logger.info(f"[EXTRACT] Success: {len(df)} rows")
        return df
    except FileNotFoundError:
        raise  # Let retry decorator handle it
    except pd.errors.ParserError as e:
        raise ExtractionError(f"CSV parsing failed: {e}") from e


def safe_transform(df):
    logger.info("[TRANSFORM] Transforming...")
    try:
        required_cols = ['employee_id', 'salary']
        missing_cols  = [c for c in required_cols if c not in df.columns]
        if missing_cols:
            raise TransformationError(f"Required columns missing: {missing_cols}")

        df = df.drop_duplicates(subset=['employee_id'])
        df['salary'].fillna(df['salary'].median(), inplace=True)

        # Data quality gate
        null_pct = df.isnull().sum().sum() / (len(df) * df.shape[1])
        if null_pct > 0.3:
            raise DataQualityError(f"Data quality check failed: {null_pct:.1%} null rate exceeds 30% threshold")

        logger.info(f"[TRANSFORM] Success: {len(df)} rows")
        return df

    except (TransformationError, DataQualityError):
        raise  # Re-raise known errors
    except Exception as e:
        raise TransformationError(f"Unexpected transform error: {e}") from e


@retry(max_attempts=3, delay_seconds=1, exceptions=(sqlite3.OperationalError,))
def safe_load(df, db='error_handling_demo.db', table='data'):
    logger.info(f"[LOAD] Writing to {db}:{table}")
    conn = None
    try:
        conn = sqlite3.connect(db)
        df_save = df.copy()
        # Convert non-serializable types
        for col in df_save.select_dtypes(include='category').columns:
            df_save[col] = df_save[col].astype(str)
        df_save.to_sql(table, conn, if_exists='replace', index=False)
        logger.info(f"[LOAD] Success: {len(df)} rows loaded")
    except sqlite3.OperationalError:
        raise  # Let retry handle DB connectivity issues
    except Exception as e:
        raise LoadError(f"Load failed: {e}") from e
    finally:
        if conn:  # Always close connection
            conn.close()
            logger.debug("[LOAD] Connection closed")


# ── Orchestrated Pipeline with Full Error Handling ─────────────
def run_safe_pipeline(filepath):
    logger.info("Pipeline starting...")
    start = time.time()

    try:
        raw   = safe_extract(filepath)
        clean = safe_transform(raw)
        safe_load(clean)
        logger.info(f"Pipeline SUCCESS in {time.time()-start:.2f}s")
        return True

    except ExtractionError as e:
        logger.error(f"Extraction stage failed: {e}")
    except TransformationError as e:
        logger.error(f"Transformation stage failed: {e}")
    except DataQualityError as e:
        logger.error(f"Data quality gate failed: {e}")
    except LoadError as e:
        logger.error(f"Load stage failed: {e}")
    except PipelineError as e:
        logger.critical(f"Unknown pipeline error: {e}")
    finally:
        elapsed = time.time() - start
        logger.info(f"Pipeline finished in {elapsed:.2f}s (success or failure)")

    return False


# ── Test: successful run ───────────────────────────────────────
print("Test 1: Successful run")
run_safe_pipeline('employees.csv')

# ── Test: file not found (triggers retry) ──────────────────────
print("\nTest 2: File not found (retry demo)")
run_safe_pipeline('nonexistent_file.csv')
````

### Q3. Different logging levels in Python. When to use each in a data pipeline context. [5 marks]

Python's `logging` module defines 5 standard severity levels (plus NOTSET):

| Level | Numeric Value | Method | When to Use in Data Pipelines |
|---|---|---|---|
| **DEBUG** | 10 | `logger.debug()` | Detailed diagnostic info: variable values, loop counts, intermediate DataFrame shapes. Only enabled during development/debugging. Never in production by default. |
| **INFO** | 20 | `logger.info()` | Normal operational milestones: "Extracted 5,000 rows", "Transformation complete", "Loaded to database". Confirms the pipeline is progressing as expected. |
| **WARNING** | 30 | `logger.warning()` | Unexpected but recoverable situations: missing values detected and imputed, duplicate rows removed, API rate limit approaching, file size larger than expected. Pipeline continues but investigation may be warranted. |
| **ERROR** | 40 | `logger.error()` | A specific task/step has failed but the pipeline may continue (e.g., one file in a batch failed to load — log error and continue with others). Requires investigation and likely manual intervention. |
| **CRITICAL** | 50 | `logger.critical()` | Catastrophic system-level failure: database connection permanently lost, disk full, configuration file corrupt. Pipeline must halt immediately. Triggers immediate PagerDuty/SMS alert. |

**Recommended Level Settings:**
```python
# Development
logging.basicConfig(level=logging.DEBUG)

# Production
logging.basicConfig(level=logging.INFO)   # File: INFO+, Console: WARNING+

# Alert integrations (e.g., Slack)
# Only send ERROR and CRITICAL to alerting channel
```

| Feature | CSV | Parquet | SQLite/PostgreSQL | MongoDB (NoSQL) |
|---|---|---|---|---|
| **Format** | Row-oriented, plain text | Columnar, binary | Row-oriented (RDBMS) | Document store (BSON/JSON) |
| **Schema** | None (inferred) | Embedded, strict | Predefined (schema-on-write) | Flexible (schema-on-read) |
| **Compression** | None | ✅ Snappy/Gzip (3–10x) | Limited | WiredTiger compression |

| **Analytical reads** | Slow (reads all columns) | ✅ Fast (reads only needed columns) | Moderate | Slow for analytical |
| **Write speed** | Fast (append) | Moderate (file write) | Moderate | Fast |
| **Joins** | Manual (pd.merge) | Spark/DuckDB | ✅ Native SQL JOIN | No native joins |

- **CSV:** Small datasets, data exchange between systems, human-readable exports, Excel-compatible deliverables. Max ~1M rows practically.

- **MongoDB:** Semi-structured or variable-schema data (IoT events, API responses), content management, real-time analytics on JSON documents, rapid iteration when schema is evolving.

### Q2. Python: Store pipeline output in SQLite, retrieve with filters, handle connection errors. [10 marks]

````python
import pandas as pd
import sqlite3
import numpy as np
import logging

logger = logging.getLogger('SQLiteStorage')
if not logger.handlers:
    logging.basicConfig(level=logging.INFO,
                        format='%(asctime)s [%(levelname)s] %(message)s')

DB_PATH = 'storage_demo.db'


# ═══════════════════════════════════════════════════════════════
# (i) STORE pipeline output in SQLite
# ═══════════════════════════════════════════════════════════════
def store_to_sqlite(df, table_name, db_path=DB_PATH, if_exists='replace'):
    """
    Store a DataFrame to SQLite with proper error handling.
    if_exists: 'replace' | 'append' | 'fail'
    """
    conn = None
    try:
        conn = sqlite3.connect(db_path)
        logger.info(f"[STORE] Connected to {db_path}")

        df.to_sql(table_name, conn, if_exists=if_exists, index=False)
        logger.info(f"[STORE] Written {len(df)} rows to table '{table_name}'")

        # Verify
        count = pd.read_sql(f'SELECT COUNT(*) AS n FROM {table_name}', conn)['n'][0]
        if count != len(df) and if_exists == 'replace':
            logger.error(f"[STORE] Row count mismatch: wrote {len(df)}, found {count}")
        else:
            logger.info(f"[STORE] Verified: {count} rows in '{table_name}'")

    except sqlite3.OperationalError as e:
        logger.error(f"[STORE] SQLite OperationalError: {e}")
        raise
    except sqlite3.DatabaseError as e:
        logger.error(f"[STORE] Database error: {e}")
        raise
    except Exception as e:
        logger.error(f"[STORE] Unexpected error: {e}", exc_info=True)
        raise
    finally:
        if conn:
            conn.close()
            logger.info("[STORE] Connection closed")


# ═══════════════════════════════════════════════════════════════
# (ii) RETRIEVE with SQL filters
# ═══════════════════════════════════════════════════════════════
def query_sqlite(sql, db_path=DB_PATH, params=None):
    """
    Execute a parameterized SQL query and return a DataFrame.
    Uses parameterized queries to prevent SQL injection.
    """
    conn = None
    try:
        conn = sqlite3.connect(db_path)
        logger.info(f"[QUERY] Executing: {sql[:80]}...")
        df = pd.read_sql(sql, conn, params=params)
        logger.info(f"[QUERY] Returned {len(df)} rows")
        return df

    except sqlite3.OperationalError as e:
        if 'no such table' in str(e):
            logger.error(f"[QUERY] Table not found. Did you run store_to_sqlite first?")
        else:
            logger.error(f"[QUERY] OperationalError: {e}")
        return pd.DataFrame()

    except Exception as e:
        logger.error(f"[QUERY] Unexpected error: {e}")
        return pd.DataFrame()
    finally:
        if conn:
            conn.close()


# ── Demo ───────────────────────────────────────────────────────
np.random.seed(42)
n = 500
sales = pd.DataFrame({
    'sale_id'  : range(1, n+1),
    'product'  : np.random.choice(['Laptop','Phone','Tablet','Watch'], n),
    'region'   : np.random.choice(['North','South','East','West'], n),
    'amount'   : np.random.uniform(100, 5000, n).round(2),
    'quantity' : np.random.randint(1, 10, n),
    'year'     : np.random.choice([2023, 2024], n)
})
sales['revenue'] = (sales['amount'] * sales['quantity']).round(2)

# Store
store_to_sqlite(sales, 'sales_data')

# Query 1: Filter by region
print("\nQuery 1: North region, 2024 sales")
q1 = query_sqlite(
    "SELECT product, SUM(revenue) as total_revenue, COUNT(*) as transactions "
    "FROM sales_data WHERE region='North' AND year=2024 "
    "GROUP BY product ORDER BY total_revenue DESC"
)
print(q1.to_string(index=False))

# Query 2: Parameterized query (safe, no SQL injection)
print("\nQuery 2: Revenue above threshold (parameterized)")
q2 = query_sqlite(
    "SELECT product, region, revenue FROM sales_data WHERE revenue > ? ORDER BY revenue DESC LIMIT 5",
    params=(40000,)
)
print(q2.to_string(index=False))

# Query 3: Aggregate summary
print("\nQuery 3: Regional summary")
q3 = query_sqlite(
    "SELECT region, COUNT(*) as sales_count, ROUND(AVG(revenue),2) as avg_revenue, "
    "ROUND(SUM(revenue),2) as total_revenue "
    "FROM sales_data GROUP BY region ORDER BY total_revenue DESC"
)
print(q3.to_string(index=False))

# Test error handling
print("\nQuery 4: Nonexistent table (error handling test)")
err = query_sqlite("SELECT * FROM nonexistent_table")
print(f"Result: {len(err)} rows (expected 0 due to error handling)")
````

### Q3. What is Parquet? Why preferred over CSV for large-scale pipeline data? How to read/write in Python? [5 marks]

**Apache Parquet** is an open-source, columnar storage file format designed for efficient analytics on large datasets. It is the de-facto standard for data lake storage in big data ecosystems (Hadoop, Spark, BigQuery, Databricks).

**Columnar vs Row Storage:**
```
CSV (row-oriented):          Parquet (columnar):
id | name  | salary          id col:     [1,2,3,...]
1  | Alice | 70000           name col:   [Alice,Bob,Carol,...]
2  | Bob   | 85000           salary col: [70000,85000,60000,...]
3  | Carol | 60000
```

For `SELECT AVG(salary) FROM employees` — CSV must read ALL columns; Parquet reads ONLY the salary column.

**Why Parquet is preferred over CSV:**

| Feature | CSV | Parquet |
|---|---|---|
| **Storage size** | Large (no compression) | 3–10× smaller (Snappy/Gzip/Zstd) |
| **Read speed (analytics)** | Slow — reads all columns | Fast — reads only needed columns |
| **Schema** | None — inferred at read time | Embedded — types preserved exactly |
| **Data types** | All stored as strings | Native int, float, datetime, bool, list |
| **Null handling** | Ambiguous empty vs null | Explicit null encoding |
| **Partitioning** | Manual folders | ✅ Native partition pruning |
| **Compatibility** | Universal | Pandas, Spark, Dask, BigQuery, Hive |

**Reading and Writing Parquet in Python:**

````python
import pandas as pd
import numpy as np
import os

# Requires: pip install pyarrow  (or fastparquet)
np.random.seed(0)
df = pd.DataFrame({
    'id'     : range(1, 10001),
    'name'   : [f'Person_{i}' for i in range(10000)],
    'salary' : np.random.randint(30000, 150000, 10000),
    'dept'   : np.random.choice(['IT','HR','Finance','Marketing'], 10000),
    'date'   : pd.date_range('2020-01-01', periods=10000, freq='H')
})

# ── Write to Parquet ───────────────────────────────────────────
df.to_parquet('output.parquet',
              engine='pyarrow',
              compression='snappy',   # Options: 'snappy', 'gzip', 'zstd', None
              index=False)

# ── Write to CSV for comparison ────────────────────────────────
df.to_csv('output.csv', index=False)

# ── Compare file sizes ─────────────────────────────────────────
parquet_size = os.path.getsize('output.parquet') / 1024
csv_size     = os.path.getsize('output.csv') / 1024
print(f"File size comparison (10,000 rows):")
print(f"  CSV:     {csv_size:.1f} KB")
print(f"  Parquet: {parquet_size:.1f} KB")
print(f"  Compression ratio: {csv_size/parquet_size:.1f}x smaller")

# ── Read Parquet — full file ───────────────────────────────────
df_full = pd.read_parquet('output.parquet', engine='pyarrow')
print(f"\nRead full parquet: {df_full.shape}")

# ── Read Parquet — specific columns only (columnar benefit) ───
df_cols = pd.read_parquet('output.parquet', columns=['id', 'salary', 'dept'])
print(f"Read selected columns only: {df_cols.shape}")
print(f"Average salary: {df_cols['salary'].mean():,.0f}")

# ── Partitioned Parquet (folder structure by dept) ─────────────
df.to_parquet('partitioned_output/',
              partition_cols=['dept'],   # Creates subfolders: dept=IT/, dept=HR/, etc.
              engine='pyarrow',
              compression='snappy')

print(f"\nPartitioned output folders:")
for item in sorted(os.listdir('partitioned_output/')):
    print(f"  partitioned_output/{item}/")

# Read partitioned — only loads IT department data
df_it = pd.read_parquet('partitioned_output/', filters=[('dept', '==', 'IT')])
print(f"\nIT department only: {len(df_it)} rows")
````

### Q4. Data partitioning in pipeline storage. How does partitioning by date/category improve query performance? [5 marks]

**Data Partitioning** is the practice of organizing data files into a hierarchical folder structure based on the values of one or more columns (partition keys), so that queries can skip irrelevant partitions entirely.

**Without partitioning:**
```
data/
  └── sales.parquet   (1TB — ALL data in one file)
```
Query `WHERE year=2024 AND month=01` must scan the entire 1TB file.

**With date partitioning:**
```
data/
  ├── year=2022/month=01/sales.parquet
  ├── year=2022/month=02/sales.parquet
  ├── year=2024/month=01/sales.parquet   ← Only this folder is read!
  └── year=2024/month=02/sales.parquet
```
Query `WHERE year=2024 AND month=01` only reads one folder — **partition pruning**.

**Performance Benefits:**
1. **Reduced I/O:** Only relevant partitions are read — for monthly data over 3 years, a single month query reads ~1/36th of total data
2. **Faster query execution:** Fewer files to open, deserialize, and scan
3. **Cost savings:** Cloud storage (S3, GCS) charges per byte scanned in query engines like Athena and BigQuery
4. **Parallel processing:** Spark can process each partition independently across workers
5. **Efficient data lifecycle:** Easy to delete old partitions (`aws s3 rm data/year=2019/ --recursive`) without touching current data

**Partitioning Best Practices:**
- Choose columns with **low-to-medium cardinality** (date, region, category) — not high-cardinality fields like `user_id` (millions of tiny files = "small file problem")
- Aim for partition files of 128MB–1GB for optimal Spark/Hive performance
- Common patterns: `year=/month=/day=/` for time series data

### Q5. What is connection pooling? Why important in data pipelines? Implement with SQLAlchemy. [5 marks]

````python
from sqlalchemy import create_engine, text, pool
import pandas as pd
import numpy as np
import threading
import time

"""
CONNECTION POOLING
==================
Creating a new database connection is expensive: TCP handshake,
authentication, session setup can take 50-200ms per connection.

In a data pipeline that executes 1000 queries per hour, creating
a new connection for each query wastes 50-200 seconds/hour in
overhead and may exhaust the DB's connection limit.

Connection Pooling maintains a POOL of reusable connections:
  1. Pool is created once at startup (N connections opened)
  2. Each query borrows a connection from the pool
  3. After the query, connection is RETURNED to pool (not closed)
  4. Next query reuses an existing connection — near-zero overhead

SQLAlchemy Pool Parameters:
  pool_size       : number of connections to maintain permanently
  max_overflow    : extra connections allowed beyond pool_size
  pool_timeout    : seconds to wait for an available connection
  pool_recycle    : seconds before connection is replaced (prevents stale)
  pool_pre_ping   : test connection health before use
"""

# ── Create Engine with Connection Pool ────────────────────────
engine = create_engine(
    'sqlite:///pool_demo.db',
    connect_args={'check_same_thread': False},   # SQLite multi-thread
    poolclass=pool.QueuePool,
    pool_size=5,         # Keep 5 connections open
    max_overflow=10,     # Allow 10 extra connections if all 5 are busy
    pool_timeout=30,     # Wait up to 30s for available connection
    pool_recycle=3600,   # Replace connections older than 1 hour
    pool_pre_ping=True   # Verify connection is alive before use
)

# ── Setup: Create and populate table ──────────────────────────
np.random.seed(0)
df = pd.DataFrame({
    'id'     : range(1, 1001),
    'value'  : np.random.randn(1000).round(4),
    'category': np.random.choice(['A','B','C'], 1000)
})
df.to_sql('metrics', engine, if_exists='replace', index=False)
print("Table 'metrics' created with 1000 rows")


# ── Query function using pooled connections ────────────────────
def query_with_pool(category):
    """Each call borrows/returns a connection from the pool."""
    with engine.connect() as conn:  # Borrows from pool
        result = pd.read_sql(
            text("SELECT category, COUNT(*) as n, AVG(value) as avg_val "
                 "FROM metrics WHERE category = :cat GROUP BY category"),
            conn, params={'cat': category}
        )
    # Connection returned to pool here (context manager __exit__)
    return result


# ── Simulate concurrent pipeline workers ──────────────────────
print("\nSimulating 10 concurrent queries using connection pool:")
results = []
start = time.time()

threads = [threading.Thread(target=lambda c=cat: results.append(query_with_pool(c)),
                            args=())
           for cat in ['A','B','C','A','B','C','A','A','B','C']]

for t in threads: t.start()
for t in threads: t.join()

print(f"  10 concurrent queries completed in {time.time()-start:.3f}s")
print(f"  Pool status: {engine.pool.checkedout()} connections in use")
print(f"  Pool status: {engine.pool.size()} connections in pool")

print("\nSample result:")
print(results[0].to_string(index=False))

# ── Bulk insert with pool ──────────────────────────────────────
new_data = pd.DataFrame({'id': range(1001, 1051), 'value': np.random.randn(50).round(4),
                         'category': np.random.choice(['A','B','C'], 50)})
new_data.to_sql('metrics', engine, if_exists='append', index=False)
print(f"\nAppended 50 rows via pooled connection")

engine.dispose()  # Close all connections in pool when done
print("Connection pool disposed.")
````

````python
import openpyxl
from openpyxl.styles import (PatternFill, Font, Alignment, Border, Side,
                              GradientFill)
from openpyxl.chart import BarChart, Reference
from openpyxl.chart.label import DataLabelList
from openpyxl.utils import get_column_letter
import pandas as pd
import numpy as np
from datetime import datetime

# ── Sample Data ────────────────────────────────────────────────
np.random.seed(42)
regions = ['North', 'South', 'East', 'West', 'Central']
data = {
    'Region'   : regions,
    'Q1_Sales' : [42, 55, 38, 61, 47],
    'Q2_Sales' : [51, 63, 44, 70, 53],
    'Q3_Sales' : [48, 59, 52, 65, 61],
    'Q4_Sales' : [65, 72, 58, 80, 69],
    'Total'    : [206, 249, 192, 276, 230]
}
df = pd.DataFrame(data)

# ── Create Workbook ────────────────────────────────────────────
wb = openpyxl.Workbook()
ws = wb.active
ws.title = 'Sales Report'

# ── (i) Title Row ──────────────────────────────────────────────
ws.merge_cells('A1:F1')
title_cell = ws['A1']
title_cell.value = f'Annual Sales Report — {datetime.now().year}'
title_cell.font      = Font(name='Calibri', size=16, bold=True, color='FFFFFF')
title_cell.fill      = PatternFill('solid', fgColor='2E4057')   # Dark navy
title_cell.alignment = Alignment(horizontal='center', vertical='center')
ws.row_dimensions[1].height = 30

# Subtitle
ws.merge_cells('A2:F2')
sub = ws['A2']
sub.value     = f'Generated: {datetime.now().strftime("%Y-%m-%d %H:%M")}  |  Unit: Lakhs NPR'
sub.font      = Font(name='Calibri', size=10, italic=True, color='666666')
sub.alignment = Alignment(horizontal='center')
ws.row_dimensions[2].height = 18

# ── Header Row ─────────────────────────────────────────────────
headers = ['Region', 'Q1 Sales', 'Q2 Sales', 'Q3 Sales', 'Q4 Sales', 'Total']
header_fill   = PatternFill('solid', fgColor='048A81')   # Teal
header_font   = Font(name='Calibri', size=11, bold=True, color='FFFFFF')
thin_border   = Border(
    left=Side(style='thin'), right=Side(style='thin'),
    top=Side(style='thin'), bottom=Side(style='thin')
)

for col_idx, header in enumerate(headers, start=1):
    cell = ws.cell(row=3, column=col_idx, value=header)
    cell.font      = header_font
    cell.fill      = header_fill
    cell.alignment = Alignment(horizontal='center', vertical='center')
    cell.border    = thin_border
ws.row_dimensions[3].height = 20

# ── (ii) Data with Alternating Row Colors ──────────────────────
fill_light = PatternFill('solid', fgColor='EAF4F4')   # Light teal
fill_white = PatternFill('solid', fgColor='FFFFFF')   # White
fill_total_col = PatternFill('solid', fgColor='FFF3CD')  # Yellow highlight for Total

for row_idx, row_data in df.iterrows():
    excel_row = row_idx + 4   # Data starts at row 4
    row_fill  = fill_light if row_idx % 2 == 0 else fill_white

    for col_idx, value in enumerate(row_data, start=1):
        cell = ws.cell(row=excel_row, column=col_idx, value=value)
        cell.fill   = fill_total_col if col_idx == 6 else row_fill
        cell.font   = Font(name='Calibri', size=10,
                           bold=(col_idx == 6))   # Bold the Total column
        cell.border = thin_border
        if col_idx > 1:  # Align numbers to right
            cell.alignment = Alignment(horizontal='right')
        else:
            cell.alignment = Alignment(horizontal='left')

# Totals row
total_row = len(df) + 4
ws.cell(row=total_row, column=1, value='TOTAL').font = Font(bold=True)
for col in range(2, 7):
    col_letter = get_column_letter(col)
    ws.cell(row=total_row, column=col,
            value=f'=SUM({col_letter}4:{col_letter}{total_row-1})')
    ws.cell(row=total_row, column=col).font = Font(bold=True)
    ws.cell(row=total_row, column=col).fill = PatternFill('solid', fgColor='048A81')
    ws.cell(row=total_row, column=col).font = Font(bold=True, color='FFFFFF')

# Column widths
col_widths = [12, 11, 11, 11, 11, 10]
for i, w in enumerate(col_widths, start=1):
    ws.column_dimensions[get_column_letter(i)].width = w

# ── (iii) Embedded Bar Chart ───────────────────────────────────
chart = BarChart()
chart.type    = 'col'
chart.title   = 'Quarterly Sales by Region'
chart.y_axis.title = 'Sales (Lakhs NPR)'
chart.x_axis.title = 'Region'
chart.style   = 10
chart.width   = 18
chart.height  = 12

# Data references (Q1-Q4 columns)
data_ref = Reference(ws, min_col=2, max_col=5, min_row=3, max_row=3+len(df))
cats_ref = Reference(ws, min_col=1, min_row=4, max_row=3+len(df))

chart.add_data(data_ref, titles_from_data=True)
chart.set_categories(cats_ref)
chart.shape = 4

ws.add_chart(chart, 'H3')

# ── Save ───────────────────────────────────────────────────────
output_path = 'sales_report.xlsx'
wb.save(output_path)
print(f"Excel report saved: {output_path}")
print(f"Features: title row, alternating row colors, totals row, embedded bar chart")
````

````python
from reportlab.lib.pagesizes import A4
from reportlab.lib.units import inch, cm
from reportlab.lib import colors
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle
from reportlab.platypus import (SimpleDocTemplate, Paragraph, Spacer, Table,
                                 TableStyle, Image as RLImage, HRFlowable)
from reportlab.lib.enums import TA_CENTER, TA_LEFT
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
import os

# ── Generate chart image for embedding ────────────────────────
np.random.seed(42)
departments = ['IT', 'HR', 'Finance', 'Marketing', 'R&D']
q_sales     = [[42,51,48,65], [28,33,30,38], [35,41,38,50], [22,27,25,32], [55,63,60,78]]

fig, ax = plt.subplots(figsize=(7, 3.5))
x = np.arange(len(departments))
quarters = ['Q1','Q2','Q3','Q4']
colors_q = ['#3498db','#2ecc71','#f39c12','#e74c3c']
width = 0.2
for i, (q, c) in enumerate(zip(quarters, colors_q)):
    ax.bar(x + i*width, [qs[i] for qs in q_sales], width, label=q, color=c, alpha=0.85)
ax.set_xticks(x + 1.5*width)
ax.set_xticklabels(departments)
ax.set_title('Quarterly Sales by Department', fontweight='bold')
ax.set_ylabel('Sales (Lakhs NPR)')
ax.legend(ncol=4, fontsize=8)
ax.spines[['top','right']].set_visible(False)
plt.tight_layout()
plt.savefig('chart_for_pdf.png', dpi=150, bbox_inches='tight')
plt.close()

# ── Build PDF ──────────────────────────────────────────────────
pdf_path = 'annual_report.pdf'
doc      = SimpleDocTemplate(pdf_path, pagesize=A4,
                              rightMargin=2*cm, leftMargin=2*cm,
                              topMargin=2*cm, bottomMargin=2*cm)

styles = getSampleStyleSheet()
elements = []

# ── Title ──────────────────────────────────────────────────────
title_style = ParagraphStyle('Title', parent=styles['Heading1'],
                              fontSize=22, spaceAfter=6,
                              textColor=colors.HexColor('#2E4057'),
                              alignment=TA_CENTER)
subtitle_style = ParagraphStyle('Subtitle', parent=styles['Normal'],
                                 fontSize=11, textColor=colors.grey,
                                 alignment=TA_CENTER, spaceAfter=20)

elements.append(Paragraph('Annual Sales Performance Report', title_style))
elements.append(Paragraph(
    f'Fiscal Year 2024 | Generated: {datetime.now().strftime("%B %d, %Y")}',
    subtitle_style
))
elements.append(HRFlowable(width='100%', thickness=2,
                            color=colors.HexColor('#048A81'), spaceAfter=15))

# ── Executive Summary Paragraph ────────────────────────────────
body_style = ParagraphStyle('Body', parent=styles['Normal'],
                             fontSize=10.5, leading=16,
                             spaceAfter=12, textColor=colors.HexColor('#2c3e50'))
section_style = ParagraphStyle('Section', parent=styles['Heading2'],
                                fontSize=13, textColor=colors.HexColor('#048A81'),
                                spaceAfter=8, spaceBefore=16)

elements.append(Paragraph('1. Executive Summary', section_style))
elements.append(Paragraph(
    'The company achieved strong revenue growth across all departments in FY2024. '
    'The R&D department led with cumulative sales of 256 lakhs NPR, representing a '
    '18% year-over-year increase. The IT department followed closely at 206 lakhs. '
    'Q4 performance was particularly strong, with all departments recording their highest '
    'quarterly figures, driven by year-end procurement cycles and successful product launches.',
    body_style
))

# ── Data Table ──────────────────────────────────────────────────
elements.append(Spacer(1, 8))
elements.append(Paragraph('2. Quarterly Sales by Department', section_style))

table_data = [['Department', 'Q1', 'Q2', 'Q3', 'Q4', 'Total']]
for dept, qs in zip(departments, q_sales):
    table_data.append([dept] + qs + [sum(qs)])
table_data.append(['TOTAL',
                   sum(r[0] for r in q_sales), sum(r[1] for r in q_sales),
                   sum(r[2] for r in q_sales), sum(r[3] for r in q_sales),
                   sum(sum(r) for r in q_sales)])

col_widths = [3.5*cm, 2*cm, 2*cm, 2*cm, 2*cm, 2.5*cm]
tbl = Table(table_data, colWidths=col_widths)
tbl.setStyle(TableStyle([
    # Header
    ('BACKGROUND',  (0,0), (-1,0),  colors.HexColor('#2E4057')),
    ('TEXTCOLOR',   (0,0), (-1,0),  colors.white),
    ('FONTNAME',    (0,0), (-1,0),  'Helvetica-Bold'),
    ('FONTSIZE',    (0,0), (-1,0),  10),
    ('ALIGN',       (0,0), (-1,-1), 'CENTER'),
    ('VALIGN',      (0,0), (-1,-1), 'MIDDLE'),
    # Alternating rows
    ('ROWBACKGROUNDS', (0,1), (-1,-2), [colors.HexColor('#EAF4F4'), colors.white]),
    # Total row
    ('BACKGROUND',  (0,-1), (-1,-1), colors.HexColor('#048A81')),
    ('TEXTCOLOR',   (0,-1), (-1,-1), colors.white),
    ('FONTNAME',    (0,-1), (-1,-1), 'Helvetica-Bold'),
    # Borders
    ('GRID',        (0,0),  (-1,-1), 0.5, colors.HexColor('#cccccc')),
    ('ROWHEIGHT',   (0,0),  (-1,-1), 22),
]))
elements.append(tbl)

# ── Embedded Chart ──────────────────────────────────────────────
elements.append(Spacer(1, 15))
elements.append(Paragraph('3. Sales Trend Visualization', section_style))
elements.append(RLImage('chart_for_pdf.png', width=14*cm, height=7*cm))

# ── Footer note ────────────────────────────────────────────────
elements.append(Spacer(1, 20))
elements.append(HRFlowable(width='100%', thickness=1, color=colors.grey))
elements.append(Paragraph(
    'Confidential — For internal use only. Auto-generated by the Analytics Pipeline. '
    f'Page generated: {datetime.now().strftime("%Y-%m-%d %H:%M")}',
    ParagraphStyle('Footer', parent=styles['Normal'],
                   fontSize=8, textColor=colors.grey, alignment=TA_CENTER)
))

# ── Build PDF ──────────────────────────────────────────────────
doc.build(elements)
print(f"PDF report saved: {pdf_path}")
print(f"File size: {os.path.getsize(pdf_path)/1024:.1f} KB")
````

### Q4. Compare Excel, HTML, and PDF for automated analytics reporting. [5 marks]

| Feature | Excel (.xlsx) | HTML | PDF |
|---|---|---|---|
| **Editability** | ✅ Fully editable by users | ✅ Editable (code) | ❌ Read-only |
| **Interactivity** | ✅ Formulas, pivot tables, charts | ✅ Interactive charts (Plotly), filters | ❌ Static |
| **Print quality** | Moderate | Poor (depends on browser) | ✅ Excellent, print-ready |
| **File size** | Medium | Small–Large (base64 charts add size) | Medium |
| **Distribution** | Email attachment | URL / web server / email | Email attachment |
| **Version control** | Difficult (binary format) | ✅ Git-friendly text | Difficult |
| **Viewer requirement** | Excel / Google Sheets | Any web browser | PDF reader |
| **Data exploration** | ✅ Sort, filter, pivot | ✅ Browser-based filtering | ❌ Fixed view |
| **Official/legal use** | ❌ | ❌ | ✅ Signature, tamper-evident |
| **Python libraries** | openpyxl, xlsxwriter | Jinja2, pandas.to_html | reportlab, pdfkit, weasyprint |

**Best format by use case:**

| Use Case | Best Format |
|---|---|
| Financial reporting for accountants who need to work with numbers | **Excel** |
| Internal analytics dashboard shared via company intranet | **HTML** |
| Regulatory compliance filing / audit document | **PDF** |
| Client-facing executive summary | **PDF** |
| Self-service data exploration for business analysts | **Excel** |
| Daily automated pipeline status report via email | **HTML** (inline) or **PDF** |

### Q5. What is Jinja2 templating? Write a Jinja2 template that renders a dynamic HTML report. [5 marks]

````python
from jinja2 import Environment, BaseLoader
import pandas as pd
import numpy as np
from datetime import datetime

"""
JINJA2 TEMPLATING
=================
Jinja2 is a fast, expressive templating engine for Python.
It allows you to embed Python-like expressions inside HTML (or any text)
using special delimiters:

  {{ variable }}          → Output a variable value
  {% statement %}         → Control flow (if/for/block)
  {# comment #}           → Comment (not rendered)

Key features:
  • Template inheritance (base templates with blocks)
  • Filters: {{ value | upper }}, {{ number | round(2) }}
  • Loops: {% for row in data %} ... {% endfor %}
  • Conditionals: {% if score > 80 %}Pass{% else %}Fail{% endif %}
  • Macros: reusable template functions
"""

# ── Sample Data ────────────────────────────────────────────────
np.random.seed(42)
students = [
    {'name': f'Student_{i}', 'math': int(np.random.randint(40,100)),
     'science': int(np.random.randint(40,100)), 'english': int(np.random.randint(40,100))}
    for i in range(1, 16)
]
for s in students:
    s['average'] = round((s['math'] + s['science'] + s['english']) / 3, 1)
    s['grade']   = 'A' if s['average'] >= 80 else ('B' if s['average'] >= 65 else ('C' if s['average'] >= 50 else 'D'))

avg_math    = round(sum(s['math'] for s in students) / len(students), 1)
avg_science = round(sum(s['science'] for s in students) / len(students), 1)
avg_english = round(sum(s['english'] for s in students) / len(students), 1)
class_avg   = round(sum(s['average'] for s in students) / len(students), 1)

# ── Jinja2 Template ────────────────────────────────────────────
template_str = """
<!DOCTYPE html>
<html>
<head>
  <title>{{ title }}</title>
  <style>
    body { font-family: Arial; margin: 30px; background: #f5f7fa; }
    h1   { color: #2E4057; border-bottom: 3px solid #048A81; padding-bottom: 8px; }
    .kpi-row { display: flex; gap: 15px; margin: 20px 0; }
    .kpi { background: #2E4057; color: white; padding: 15px 20px;
           border-radius: 8px; text-align: center; flex: 1; }
    .kpi h3 { font-size: 12px; opacity: 0.8; }
    .kpi .val { font-size: 28px; font-weight: bold; }
    table { width: 100%; border-collapse: collapse; background: white;
            border-radius: 8px; overflow: hidden;
            box-shadow: 0 2px 8px rgba(0,0,0,.08); }
    th { background: #048A81; color: white; padding: 12px; text-align: left; }
    td { padding: 10px 12px; border-bottom: 1px solid #ecf0f1; }
    tr:nth-child(even) td { background: #f8fbfb; }
    .grade-A { color: #27ae60; font-weight: bold; }
    .grade-B { color: #2980b9; font-weight: bold; }
    .grade-C { color: #f39c12; font-weight: bold; }
    .grade-D { color: #e74c3c; font-weight: bold; }
    .footer { margin-top: 30px; color: #95a5a6; font-size: 12px; text-align: center; }
  </style>
</head>
<body>
  <h1>{{ title }}</h1>
  <p style="color:#7f8c8d;">Generated: {{ generated_at }} | Total Students: {{ students | length }}</p>

  {# KPI tiles #}
  <div class="kpi-row">
    <div class="kpi"><h3>Class Average</h3><div class="val">{{ class_avg }}</div></div>
    <div class="kpi"><h3>Avg Math</h3><div class="val">{{ avg_math }}</div></div>
    <div class="kpi"><h3>Avg Science</h3><div class="val">{{ avg_science }}</div></div>
    <div class="kpi"><h3>Avg English</h3><div class="val">{{ avg_english }}</div></div>
    <div class="kpi"><h3>Pass Rate</h3>
      <div class="val">{{ (students | selectattr('average', 'ge', 50) | list | length / students | length * 100) | round(0) | int }}%</div>
    </div>
  </div>

  <h2 style="color:#2E4057;margin-bottom:12px;">Student Results</h2>
  <table>
    <thead>
      <tr><th>Name</th><th>Math</th><th>Science</th><th>English</th><th>Average</th><th>Grade</th></tr>
    </thead>
    <tbody>
      {% for student in students %}
      <tr>
        <td>{{ student.name }}</td>
        <td>{{ student.math }}</td>
        <td>{{ student.science }}</td>
        <td>{{ student.english }}</td>
        <td><strong>{{ student.average }}</strong></td>
        <td class="grade-{{ student.grade }}">{{ student.grade }}</td>
      </tr>
      {% endfor %}
    </tbody>
  </table>

  <div class="footer">Auto-generated by Jinja2 | IOE Data Analytics Pipeline</div>
</body>
</html>
"""

# ── Render Template ────────────────────────────────────────────
env      = Environment(loader=BaseLoader())
template = env.from_string(template_str)

rendered_html = template.render(
    title        = 'Student Performance Report — Class 12',
    generated_at = datetime.now().strftime('%B %d, %Y at %H:%M'),
    students     = students,
    class_avg    = class_avg,
    avg_math     = avg_math,
    avg_science  = avg_science,
    avg_english  = avg_english
)

with open('jinja2_report.html', 'w', encoding='utf-8') as f:
    f.write(rendered_html)

print(f"Jinja2 HTML report saved: jinja2_report.html")
print(f"Students rendered: {len(students)}")
print(f"File size: {len(rendered_html)/1024:.1f} KB")
````

### Q6. Automate distribution of reports via email using smtplib. [5 marks]

````python
import smtplib
import os
from email.mime.multipart import MIMEMultipart
from email.mime.text     import MIMEText
from email.mime.base     import MIMEBase
from email               import encoders
from datetime            import datetime

"""
EMAIL REPORT DISTRIBUTION
==========================
Automate sending of generated Excel/PDF/HTML reports to stakeholders.
Uses Python's built-in smtplib + email.mime libraries.

Security best practices:
  1. Store credentials in environment variables (never hardcode)
  2. Use App Passwords for Gmail (not your main password)
  3. Enable TLS (STARTTLS on port 587 or SSL on port 465)
  4. For production: use SendGrid / AWS SES for reliable delivery
"""


def send_report_email(
    to_addresses,
    subject,
    body_html,
    attachment_paths=None,
    smtp_host='smtp.gmail.com',
    smtp_port=587
):
    """
    Send an email with optional file attachments via Gmail SMTP.

    Parameters:
        to_addresses     : list of recipient email addresses
        subject          : email subject line
        body_html        : HTML body content
        attachment_paths : list of file paths to attach
        smtp_host        : SMTP server (default: Gmail)
        smtp_port        : SMTP port (587 for TLS, 465 for SSL)
    """
    # ── Credentials from environment variables ─────────────────
    sender    = os.environ.get('EMAIL_SENDER', 'pipeline@company.com')
    password  = os.environ.get('EMAIL_PASSWORD', 'your-app-password')

    # ── Build message ──────────────────────────────────────────
    msg = MIMEMultipart('mixed')
    msg['From']    = sender
    msg['To']      = ', '.join(to_addresses)
    msg['Subject'] = subject

    # HTML body
    msg.attach(MIMEText(body_html, 'html'))

    # ── Attach files ───────────────────────────────────────────
    if attachment_paths:
        for filepath in attachment_paths:
            if not os.path.exists(filepath):
                print(f"  WARNING: Attachment not found: {filepath}")
                continue

            with open(filepath, 'rb') as f:
                part = MIMEBase('application', 'octet-stream')
                part.set_payload(f.read())

            encoders.encode_base64(part)
            part.add_header(
                'Content-Disposition',
                f'attachment; filename="{os.path.basename(filepath)}"'
            )
            msg.attach(part)
            print(f"  Attached: {os.path.basename(filepath)} "
                  f"({os.path.getsize(filepath)/1024:.1f} KB)")

    # ── Send via SMTP ──────────────────────────────────────────
    try:
        server = smtplib.SMTP(smtp_host, smtp_port)
        server.ehlo()
        server.starttls()                    # Enable TLS encryption
        server.login(sender, password)
        server.sendmail(sender, to_addresses, msg.as_string())
        server.quit()
        print(f"  Email sent successfully to: {', '.join(to_addresses)}")
        return True

    except smtplib.SMTPAuthenticationError:
        print("  ERROR: Authentication failed. Check EMAIL_SENDER/EMAIL_PASSWORD env vars.")
    except smtplib.SMTPConnectError:
        print(f"  ERROR: Cannot connect to {smtp_host}:{smtp_port}")
    except Exception as e:
        print(f"  ERROR: {e}")
    return False


# ── Demo: Build email body and call function ───────────────────
email_body = f"""
<html><body>
<h2 style="color:#2E4057;">📊 Daily Analytics Report</h2>
<p>Dear Team,</p>
<p>Please find attached the automated analytics report for
<strong>{datetime.now().strftime('%B %d, %Y')}</strong>.</p>
<table style="border-collapse:collapse;">
  <tr><td style="padding:6px 12px;">📁 Excel Report:</td><td>sales_report.xlsx</td></tr>
  <tr><td style="padding:6px 12px;">📄 PDF Summary:</td><td>annual_report.pdf</td></tr>
</table>
<p style="margin-top:15px;color:#95a5a6;font-size:12px;">
  This is an automated email. Do not reply directly.<br>
  Generated by the Data Analytics Pipeline | IOE BE Computer Engineering
</p>
</body></html>
"""

print("Email Distribution Demo:")
print("  Recipient: manager@company.com")
print(f"  Subject: Daily Analytics Report — {datetime.now().strftime('%Y-%m-%d')}")
print("  Attachments: sales_report.xlsx, annual_report.pdf")
print()
print("NOTE: To actually send, set environment variables:")
print("  export EMAIL_SENDER='your@gmail.com'")
print("  export EMAIL_PASSWORD='your-16-char-app-password'")
print("Then call:")
print("""
  send_report_email(
      to_addresses    = ['manager@company.com'],
      subject         = f'Daily Report — {datetime.now().strftime("%Y-%m-%d")}',
      body_html       = email_body,
      attachment_paths= ['sales_report.xlsx', 'annual_report.pdf']
  )
""")
````

### Q2. Implement a mini end-to-end automated pipeline: CSV → clean/aggregate → SQLite → HTML report → scheduled daily. [10 marks]

````python
import pandas as pd
import numpy as np
import sqlite3
import schedule
import time
import logging
import os
from datetime import datetime
from jinja2 import Environment, BaseLoader

# ── Setup ──────────────────────────────────────────────────────
os.makedirs('pipeline_outputs', exist_ok=True)
os.makedirs('logs', exist_ok=True)

logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s [%(levelname)s] %(message)s',
    handlers=[
        logging.FileHandler('logs/full_pipeline.log', mode='a'),
        logging.StreamHandler()
    ]
)
logger = logging.getLogger('FullPipeline')


# ═══════════════════════════════════════════════════════════════
# STEP 1: EXTRACT
# ═══════════════════════════════════════════════════════════════
def extract(csv_path='raw_sales.csv'):
    logger.info(f"[1/5 EXTRACT] Reading {csv_path}")
    if not os.path.exists(csv_path):
        raise FileNotFoundError(f"Source file not found: {csv_path}")
    df = pd.read_csv(csv_path)
    logger.info(f"[1/5 EXTRACT] Loaded {len(df)} rows")
    return df


# ═══════════════════════════════════════════════════════════════
# STEP 2: TRANSFORM & AGGREGATE
# ═══════════════════════════════════════════════════════════════
def transform(df):
    logger.info("[2/5 TRANSFORM] Cleaning and aggregating...")

    # Clean
    df = df.drop_duplicates(subset=['order_id'])
    df['unit_price'].fillna(df['unit_price'].median(), inplace=True)
    df['quantity'].fillna(1, inplace=True)
    df['quantity']   = df['quantity'].astype(int)
    df['order_date'] = pd.to_datetime(df['order_date'], format='%d/%m/%Y', errors='coerce')
    df = df[df['status'] == 'completed'].copy()

    # Enrich
    df['revenue']     = (df['quantity'] * df['unit_price']).round(2)
    df['order_month'] = df['order_date'].dt.to_period('M').astype(str)

    # Aggregate: product-level summary
    summary = df.groupby('product').agg(
        total_orders   = ('order_id', 'count'),
        total_revenue  = ('revenue', 'sum'),
        avg_order_value= ('revenue', 'mean'),
        total_units    = ('quantity', 'sum')
    ).round(2).reset_index().sort_values('total_revenue', ascending=False)

    logger.info(f"[2/5 TRANSFORM] Done. {len(df)} orders → {len(summary)} product groups")
    return df, summary


# ═══════════════════════════════════════════════════════════════
# STEP 3: LOAD TO SQLITE
# ═══════════════════════════════════════════════════════════════
def load(df_raw, df_summary, db='pipeline_outputs/pipeline.db'):
    logger.info(f"[3/5 LOAD] Writing to SQLite: {db}")
    conn = sqlite3.connect(db)
    try:
        df_save = df_raw.copy()
        df_save['order_date'] = df_save['order_date'].astype(str)
        df_save.to_sql('sales_transactions', conn, if_exists='replace', index=False)
        df_summary.to_sql('product_summary', conn, if_exists='replace', index=False)
        logger.info(f"[3/5 LOAD] Stored {len(df_raw)} transactions + {len(df_summary)} summaries")
    finally:
        conn.close()


# ═══════════════════════════════════════════════════════════════
# STEP 4: GENERATE HTML REPORT
# ═══════════════════════════════════════════════════════════════
REPORT_TEMPLATE = """
<!DOCTYPE html><html><head><title>Sales Report</title>
<style>
  body{font-family:Arial;margin:25px;background:#f5f7fa;}
  h1{color:#2E4057;border-bottom:3px solid #048A81;padding-bottom:8px;}
  .kpi-row{display:flex;gap:12px;margin:18px 0;}
  .kpi{background:#2E4057;color:white;padding:14px 18px;border-radius:8px;flex:1;text-align:center;}
  .kpi h3{font-size:11px;opacity:.8;}
  .kpi .val{font-size:26px;font-weight:700;}
  table{width:100%;border-collapse:collapse;background:white;border-radius:8px;
        box-shadow:0 2px 8px rgba(0,0,0,.08);}
  th{background:#048A81;color:white;padding:11px;text-align:left;}
  td{padding:9px 11px;border-bottom:1px solid #ecf0f1;}
  tr:nth-child(even) td{background:#f8fbfb;}
  .footer{text-align:center;color:#95a5a6;font-size:11px;margin-top:25px;}
</style></head><body>
<h1>📊 Daily Sales Report</h1>
<p style="color:#7f8c8d;">Generated: {{ run_time }} | Source: {{ source }}</p>
<div class="kpi-row">
  <div class="kpi"><h3>Total Orders</h3><div class="val">{{ total_orders }}</div></div>
  <div class="kpi"><h3>Total Revenue</h3><div class="val">₹{{ total_revenue }}</div></div>
  <div class="kpi"><h3>Top Product</h3><div class="val">{{ top_product }}</div></div>
  <div class="kpi"><h3>Products</h3><div class="val">{{ num_products }}</div></div>
</div>
<h2 style="color:#2E4057;margin:20px 0 10px;">Product Performance Summary</h2>
<table>
  <tr><th>Product</th><th>Orders</th><th>Total Revenue (₹)</th><th>Avg Order</th><th>Units Sold</th></tr>
  {% for row in rows %}
  <tr>
    <td><strong>{{ row.product }}</strong></td>
    <td>{{ row.total_orders }}</td>
    <td>{{ '{:,.2f}'.format(row.total_revenue) }}</td>
    <td>{{ '{:,.2f}'.format(row.avg_order_value) }}</td>
    <td>{{ row.total_units }}</td>
  </tr>
  {% endfor %}
</table>
<div class="footer">Auto-generated | Data Analytics Pipeline | IOE BE Computer Engineering</div>
</body></html>
"""

def generate_report(df_raw, df_summary):
    logger.info("[4/5 REPORT] Generating HTML report...")
    env      = Environment(loader=BaseLoader())
    template = env.from_string(REPORT_TEMPLATE)

    html = template.render(
        run_time     = datetime.now().strftime('%Y-%m-%d %H:%M:%S'),
        source       = 'raw_sales.csv',
        total_orders = len(df_raw),
        total_revenue= f"{df_raw['revenue'].sum():,.2f}",
        top_product  = df_summary.iloc[0]['product'],
        num_products = len(df_summary),
        rows         = df_summary.to_dict('records')
    )

    report_path = f"pipeline_outputs/report_{datetime.now().strftime('%Y%m%d_%H%M%S')}.html"
    with open(report_path, 'w') as f:
        f.write(html)
    logger.info(f"[4/5 REPORT] Saved: {report_path}")
    return report_path


# ═══════════════════════════════════════════════════════════════
# STEP 5: FULL PIPELINE RUNNER
# ═══════════════════════════════════════════════════════════════
def run_pipeline():
    start = datetime.now()
    logger.info(f"{'='*55}")
    logger.info(f"PIPELINE STARTED: {start.strftime('%Y-%m-%d %H:%M:%S')}")
    try:
        raw_df                  = extract('raw_sales.csv')
        clean_df, summary_df    = transform(raw_df)
        load(clean_df, summary_df)
        report_path             = generate_report(clean_df, summary_df)

        elapsed = (datetime.now() - start).total_seconds()
        logger.info(f"[5/5] PIPELINE SUCCESS in {elapsed:.2f}s | Report: {report_path}")
        return True
    except Exception as e:
        logger.error(f"PIPELINE FAILED: {e}", exc_info=True)
        return False


# ── SCHEDULING ──────────────────────────────────────────────────
# Production: schedule.every().day.at('08:00').do(run_pipeline)
# Demo: run once directly

logger.info("Running pipeline directly (demo mode)")
result = run_pipeline()

print(f"\nPipeline {'SUCCESS ✅' if result else 'FAILED ❌'}")
print("Reports saved in: pipeline_outputs/")
print("\nTo schedule daily at 8 AM, add to your script:")
print("  schedule.every().day.at('08:00').do(run_pipeline)")
print("  while True:")
print("      schedule.run_pending()")
print("      time.sleep(60)")
````

### Q3. Key challenges in building a production-grade automated data pipeline. [5 marks]

**1. Fault Tolerance**
- **Challenge:** Any step can fail — network timeouts, disk full, source schema changes, memory errors
- **Solutions:** Automatic retry with exponential backoff; dead-letter queues for failed records; idempotent operations so re-runs don't corrupt data; circuit breakers to stop cascading failures

**2. Scalability**
- **Challenge:** Data volumes grow over time; what works for 1GB fails at 1TB
- **Solutions:** Data partitioning (process in chunks); distributed processing (Apache Spark); horizontal scaling (parallel workers); incremental loads instead of full reprocessing

**3. Data Freshness (SLA)**
- **Challenge:** Business needs data by 8 AM; pipeline may run late due to upstream delays or large data volumes
- **Solutions:** SLA monitoring with alerts; source data availability checks before starting; pipeline parallelization; priority queues for critical paths

**4. Observability**
- **Challenge:** Complex multi-step pipelines are black boxes — hard to debug failures or understand why output changed
- **Solutions:** Structured logging at every step; metrics (row counts, null rates, execution time) stored in monitoring DB; distributed tracing with `trace_id`; data lineage tracking; Airflow/Dagster web UI dashboards

**5. Schema Evolution**
- **Challenge:** Source systems add/rename/remove columns without notice, breaking downstream transformations
- **Solutions:** Schema validation on ingestion (Great Expectations, Pandera); schema registry (Apache Avro + Schema Registry); alert on unexpected schema change rather than silently proceeding

**6. Data Quality at Scale**
- **Challenge:** Small quality issues (0.1% nulls) are acceptable in samples but represent thousands of bad records at scale
- **Solutions:** Automated data quality gates with configurable thresholds; row count reconciliation between source and target; automated anomaly detection on key metrics

### Q4. How to test and validate an automated data pipeline before deploying to production? [5 marks]

**Testing Strategy for Data Pipelines:**

**1. Unit Testing (pytest)**
Test each pipeline function in isolation with known inputs and expected outputs:
```python
def test_transform_removes_duplicates():
    df = pd.DataFrame({'id': [1,1,2], 'val': [10,10,20]})
    result = transform(df)
    assert len(result) == 2, "Duplicates not removed"

def test_transform_fills_nulls():
    df = pd.DataFrame({'id': [1,2], 'salary': [50000, None]})
    result = transform(df)
    assert result['salary'].isnull().sum() == 0, "Nulls not filled"
```

**2. Integration Testing**
Run the full pipeline end-to-end on a small representative test dataset:
- Use a test database (SQLite in memory: `sqlite:///:memory:`)
- Verify: rows loaded = expected count, output schema matches expectation, SQL queries return valid results

**3. Data Validation with Great Expectations**
Define and enforce data quality expectations:
```python
# Expectations defined once, validated on every run
df.expect_column_values_to_not_be_null('order_id')
df.expect_column_values_to_be_between('price', min_value=0, max_value=100000)
df.expect_column_values_to_be_in_set('status', ['completed','pending','cancelled'])
```

**4. Schema Validation (Pandera)**
```python
import pandera as pa
schema = pa.DataFrameSchema({'price': pa.Column(float, pa.Check.ge(0)),
                              'status': pa.Column(str)})
schema.validate(df)  # Raises SchemaError if violated
```

**5. Regression Testing**
Capture "golden output" from a known-good run; compare new output against it to detect unexpected behavior changes.

**6. Load/Performance Testing**
Test with 10× production data volume before go-live to identify bottlenecks and memory issues.

| Tool/Practice | Role |
|---|---|
| **GitHub Actions / GitLab CI** | Automate test runs on every commit |
| **dbt (data build tool)** | Version-controlled SQL transformations with built-in tests; `dbt test` validates data contracts |
| **Docker** | Containerize pipeline code — ensures "works on my machine" = works in production |
| **Airflow + Git sync** | DAG files version-controlled in Git; automatically picked up by Airflow on deploy |
| **Schema Registry** | Version database schemas; migrations tracked with Alembic (SQLAlchemy) or Flyway |
| **Feature flags** | Deploy new pipeline version in shadow mode (runs but output not used) before full cutover |
| **Blue-green deployment** | Run old and new pipeline in parallel; switch traffic when new version is validated |

**Benefits:** Eliminates "it worked in dev but broke in prod" scenarios; provides automated safety net; enables confident, frequent deployments; creates audit trail of every change.
