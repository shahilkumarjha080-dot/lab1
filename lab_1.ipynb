# Basic data types:
    Python's data types are the basic building blocks of any scripts or application. Python is dynamically typed hence interpreter figures it out based on the value assigned. 
Most common build in data types:

1. Numerical Types:
    Used to represent numbers.
    
    i. integer (int) : Whole numbers, positive, negative.
    - examples: x=5 , y= -5
    
    ii. Float(float) : Numbers with decimal point or exponential form.
    - examples: x=3.14

    iii. complex(complex): Numebers with a real and imaginary part 
    - examples: x= 3+5j

2. Sequesnce Types:
    Used to store collection of data.

    i. String (str): A sequence of Unicode characters wrapped in quotes.
    - examples: "Hello world"

    ii. List(list): An ordered and mutable collection.
    - examples: [1, "apple", 3.5]

    iii. Tuple(tuple): An ordered and immutable collection.
    - exaomples: (10,20,30)

3. Mapping and Set Types
    used for unique data and key-values relationships.

    i. Dictorary(dict): A collection of key values pairs. 
    - examples: {"name": "Ram", "age": 20}

    ii. Set (set): An unordered collection of unique items.
    -examples: {1,2,3,3} becomes {1,2,3}

4. Others 

    i. Boolean(bool): Represent logical values true or false
    - examples: is_active= TRUE  

    ii. None type(NoneType): Represents the absence of a value or a null value. 
    - examples: x= None



## The type() function
    The type() function is a build-in utility used to identify the class (data type) of an object. 

- Implementation and Working:
    In python, everything is an object. When you pass a variable to type(), the interpreter looks at the object's internal metadata to see which class it was instantinated from.

- Working mechanism:
    1. You pass an object (or variable) as an argument: type(object).
    2. The function returns a type object, typically desplayed as <class 'type_name'>.
    3. It is most commonly used for debugging or ensuring a variable contains the expected data before performing an operation. 




```python
age=25 
print (type(age))
```

## Need for typecasting:
    Since input() always returns a string, we often need Typecasting-the process of converting one data type into another. Without this the mathematical operations would fail. 


##  Program: Adding Two Numbers


```python
def calculate_area(length, width):
    area = length * width
    return area

l = float(input("Enter length: "))
w = float(input("Enter width: "))

result = calculate_area(l, w)
print("length is:", l)
print("width is:", w)
print(f"the area of the rectangle is: {result}")
```
